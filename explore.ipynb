```python
import pickle
import pandas as pd

split = ['S04', 'S07', 'S08']
final_list=[]
for i in range(len(split)):
    featuresRGB = pd.read_pickle('raw_features/'+ split[i]+'_RGB_features.pkl')
    featuresEMG = pd.read_pickle('raw_features/'+ split[i]+'_EMG_features.pkl')
    for elem1 in featuresRGB['features']:
        for elem2 in featuresEMG['features']:
            if elem1['uid'] == elem2['uid']:
                elem1['features_EMG'] = elem2['features_EMG']
                elem1['label'] = elem2['label']
                elem1['label_code'] = elem2['label_code']
                elem1['uid'] = split[i]+'_'+str(elem1['uid'])
                final_list.append(elem1)
    print('DONE'+split[i])

for i in range(len(final_list)):
    final_list[i]['uid_complete']=final_list[i]['uid']
    final_list[i]['uid'] = i
file = open('saved_features/featuresS04S07S08.pkl','wb')
output = {}
output['features'] = final_list
print(len(output['features']))
pickle.dump(output,file)
print('DONE')
```

```python
import random
import pickle
import pandas as pd
l_tot = 1142
l_test = 343

test_indexes = random.sample(range(l_tot),l_test)

test_annotations = pd.DataFrame(columns=['uid', 'label', 'verb_class'])
train_annotations = pd.DataFrame(columns=['uid', 'label', 'verb_class'])
features = pd.read_pickle('saved_features/features_S04S07S08.pkl')
for i in test_indexes:
    elem = features['features'][i]
    test_annotations.loc[len(test_annotations)] = [elem['uid'], elem['label'], elem['label_code']]

for j in range(l_tot):
    if j not in test_indexes:
        elem = features['features'][j]
        train_annotations.loc[len(train_annotations)] = [elem['uid'], elem['label'], elem['label_code']]

file = open('annotations1/S04S07S08_test.pkl','wb')
pickle.dump(test_annotations,file)
file.close()

file = open('annotations1/S04S07S08_train.pkl','wb')
pickle.dump(train_annotations,file)
file.close()

print('DONE')
```
